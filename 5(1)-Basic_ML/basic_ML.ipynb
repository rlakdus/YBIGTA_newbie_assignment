## 1. 데이터 로드 및 기본 탐색

```python
import pandas as pd
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
print(df["Class"].value_counts())
print(df["Class"].value_counts(normalize=True))
```

## 2. 샘플링

```python
class1 = df[df["Class"]==1]
class0 = df[df["Class"]==0].sample(n=10000,random_state=42)
df_new = pd.concat([class1, class0], axis=0)

print(df_new["Class"].value_counts())
print(df_new["Class"].value_counts(normalize=True))
```

```python
df_new
```

## 3. 데이터 전처리

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_new["Amount_Scaled"] = scaler.fit_transform(df_new[["Amount"]])
df_new.drop(columns=["Amount"], inplace=True)

df_new.head()
```

```python
X = df_new.drop(columns=["Class"])
y = df_new["Class"]

print(X.shape, y.shape)
```

## 4. 학습 데이터와 테스트 데이터 분할

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

## 5. SMOTE 적용

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

print("Before SMOTE")
print(y_train.value_counts())
print("After SMOTE")
print(y_train_sm.value_counts())
```

## 6. 모델 학습

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000,
    class_weight=None,  # SMOTE를 쓰므로 불필요
    random_state=42
)

model.fit(X_train_sm, y_train_sm)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_proba = model.predict_proba(X_test)[:, 1]
print(y_proba)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import average_precision_score

pr_auc = average_precision_score(y_test, y_proba)
print("PR-AUC:", pr_auc)
```

```python
threshold = 0.7
y_pred_thresh = (y_proba >= threshold).astype(int)
y_pred_thresh


print(classification_report(y_test, y_pred_thresh))
```
